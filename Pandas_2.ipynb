### **Convert a date column from object datatype to datetime datatype**

```python
date = pd.to_datetime(df['order_date'])
date
```

```python
df['converted_order_date'] = pd.to_datetime(df['order_date'])
df
```

```python
df.dtypes
```

### **Extract year from new created date column and create a new column for year**

```python
df['order_date_year'] = df['converted_order_date'].dt.year
df.head()
```

```python
df['order_date_month'] = df['converted_order_date'].dt.month
df.head()
```

```python
df['order_date_week'] = df['converted_order_date'].dt.weekofyear
df.head()
```

The week is showing 52 because it considers the week 1 from the december month becasue we have executed week of the year.

```python
df[df['order_date_year'] == 2014]
```

### **In which month we are doing the maximum sale**

```python
df['order_date_month'].value_counts()
```

Here in value_counts() we can say that in 11th month there are highest number of transactions so the sales will be more in that month.
